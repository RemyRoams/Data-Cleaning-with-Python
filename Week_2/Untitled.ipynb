```python
import pandas as pd
```

```python
# Importing data
df_flow = pd.read_csv('PD 2024 Wk 1 Output Flow Card.csv')
df_flow.head()
```

```python
# Importing data
df_noflow = pd.read_csv('PD 2024 Wk 1 Output Non-Flow Card.csv')
df_noflow.head()
```

```python
# Union of the two dataframes
df = pd.concat([df_flow, df_noflow], ignore_index = True)
df
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
df["Quarter"] = df["Date"].dt.quarter
df = df.drop(columns=["Date"])
df
```

```python
df.groupby('Quarter,Flow Card?,Class')['Price'].median()
df.groupby('Quarter,Flow Card?,Class')['Price'].median()
df.groupby('Quarter,Flow Card?,Class')['Price'].median()
```

```python
df.groupby('Quarter,Flow Card?,Class')['Price'].min()
df.groupby('Quarter,Flow Card?,Class')['Price'].min()
df.groupby('Quarter,Flow Card?,Class')['Price'].min()
```

```python
df.groupby('Quarter,Flow Card?,Class')['Price'].max()
df.groupby('Quarter,Flow Card?,Class')['Price'].max()
df.groupby('Quarter,Flow Card?,Class')['Price'].max()
```

```python
df.groupby(['Quarter', 'Flow Card?', 'Class'])['Price'].median().reset_index()
df.groupby(['Quarter', 'Flow Card?', 'Class'])['Price'].min().reset_index()
df.groupby(['Quarter', 'Flow Card?', 'Class'])['Price'].max().reset_index()
```
